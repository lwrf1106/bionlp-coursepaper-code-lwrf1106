```python
import requests
import json
import time
import os
import re
```

```python
with open('search.results.litcovid.tsv', 'r') as f:
    lines = f.read().splitlines()
```

```python
lines = lines[34:]
```

```python
pid_title = {}
for line in lines:
    pid, title, _ = line.split('\t')
    pid_title[pid] = title
    # pid denotes pubmed id
```

```python
# batch download
batch_size = 50
pid_list = list(pid_title.keys())
pidhash_pidstr = {}
for i in range(0, len(pid_title), batch_size):
    batch_pids = pid_list[i:i+batch_size]
    for pid in batch_pids.copy():
        if os.path.exists(f'raw_data/{pid}.txt'):
            print(f'{pid} already exists')
            batch_pids.remove(pid)
    if batch_pids:
        pid_str = ','.join(batch_pids)
        pid_hash = hash(pid_str)
        url = f'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed&id={pid_str}&rettype=abstract&retmode=text'
        r = requests.get(url)
        with open(f'batched_download/{pid_hash}.txt', 'w') as f:
            f.write(r.text)
        print(f'{pid_hash} downloaded')
        pidhash_pidstr[pid_hash] = pid_str
```

```python
with open('pid_title.json', 'w') as f:
    json.dump(pid_title, f)
```

```python
with open('pidhash_pidstr.json', 'w') as f:
    json.dump(pidhash_pidstr, f)
```

```python
# split batch file
for pid_hash, pid_str in pidhash_pidstr.items():
    pids = pid_str.split(',')
    with open(f'batched_download/{pid_hash}.txt', 'r') as f:
        lines = f.read().splitlines()
    begin_line_indexs = [0]
    end_line_indexs = []
    end_flag = False
    for i, line in enumerate(lines):
        if line:
            end_flag = False
        else:
            if end_flag:
                end_line_indexs.append(i)
                begin_line_indexs.append(i + 1)
                end_flag = False
            else:
                end_flag = True
    end_line_indexs.append(len(lines))
    assert len(begin_line_indexs) == len(end_line_indexs)

    for index, (begin_idx, end_idx) in enumerate(zip(begin_line_indexs, end_line_indexs)):
        pid = pids[index]
        text = '\n'.join(lines[begin_idx:end_idx+1]).strip()
        pid_extract = re.findall(r'PMID: (\d+)', text)
        assert pid_extract == [pid]
        with open(f'raw_data/{pid}.txt', 'w') as f:
            f.write(text)
        print(f'{pid} saved')
```
